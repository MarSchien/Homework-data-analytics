```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
fileToLoad = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchaseData = pd.read_csv(fileToLoad)
```

## Player Count

* Display the total number of players


```python
# purchaseData.head()
# count = purchaseData["Purchase ID"].value_counts()

unique = purchaseData['SN'].unique()

totalPlayer = len(unique)
totalPlayer

totalPlayerDf = pd.DataFrame({'Total players': [totalPlayer]})
totalPlayerDf
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uniqueItems1 = purchaseData['Item ID'].unique()
uniqueItems2 = len(uniqueItems1)

averagePrice = purchaseData['Price'].mean()
totalRevenue = purchaseData['Price'].sum()
totalRows = len(purchaseData.index)

purchaseAnalysisDf = pd.DataFrame({'Number of Unique Items': [uniqueItems2], 'Average Price': [averagePrice], 
                                   'Number of Purchases': [totalRows], 'Total Revenue': [totalRevenue]})

purchaseAnalysisDf["Average Price"] = purchaseAnalysisDf["Average Price"].map("${:.2f}".format)
purchaseAnalysisDf["Total Revenue"] = purchaseAnalysisDf["Total Revenue"].map("{:,}".format)

purchaseAnalysisDf 
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

twoColumns = purchaseData.loc[:, ['Gender','SN']]

dropSamePeople = twoColumns.drop_duplicates(["SN"])

genderCounts = dropSamePeople['Gender'].value_counts()

genderCountsDf = pd.DataFrame(genderCounts)

totalGender = genderCountsDf['Gender'].sum()

percentage = (genderCountsDf['Gender']/totalGender)*100

genderCountsDf.dtypes

genderCountsDf['Percentage of Players'] = percentage

genderCountsDf['Percentage of Players'] = genderCountsDf['Percentage of Players'].map("{:.2f}%".format)

genderCountsDf


```
